# CREATE IMAGE PATCHES

In this notebook the main goal is to create, from training dataset, a huge amount of different patches relative to different lichens. 

For each image, we will crop a central subimage, from which we will extract a fixed number of image patches of dimension (200,200,3).

```python
# create image patches 
pth = "../../final_dataset/train"
test_pth = "../../final_dataset/test"

# cosa creaiamo: patches e le salviamo in file pickle. label e salviamo in file pickle 
# file csv dove salviamo nome patch (nome_lichene + texture + numero di patch) e sua classificazione 
# patches in cartelle 


def extract_image_patches(pth,patch_dimension = 100,res_dim = 1000,crop_f = 100, patch_dir = "patches/train"):
    
    if(os.path.exists("patches/")==False):
        os.mkdir("patches/")
    #create patches directory
    if (os.path.exists(patch_dir)==False):
        os.mkdir(patch_dir)
    
    x_dims = np.arange(0,1000,200)
    y_dims = np.arange(0,1000,200) 
    
    lichens = os.listdir(pth)
    df = pd.DataFrame(columns = ['patch', 'directory', 'lichen','class'])
    
    features = []
    label = []
    digit_indices = {}
    dic_label = {}
    for i,l in enumerate(lichens):
        print("+++++")
        print(l)
        immagini = []
        if(l == ".DS_Store"):
            continue
        #create sepcific_lichens_folder 
        lichen_folder = os.path.join(patch_dir,l)
        if(os.path.exists(lichen_folder)==False):
            os.mkdir(lichen_folder)


        lichens_path = os.path.join(pth,l)
        dic_label[i] = l

        # lista_di_immagini
        specific_lichen_images_list = os.listdir(lichens_path)

        for j,images in enumerate(specific_lichen_images_list):
            if(".DS_Store" in images):
                continue 

            # work on images 
            img_path = os.path.join(lichens_path,images)
            print(img_path)
            im = cv2.imread(img_path)
            im = cv2.resize(im,(res_dim,res_dim))

            #take central crop 
            im = im[crop_f:res_dim - crop_f,crop_f:res_dim - crop_f,:]


            crop_count = 1
            for ii in range(1,len(x_dims)):
                for jj in range(1,len(y_dims)):
                    crop_img = im[x_dims[ii -1]:x_dims[ii],y_dims[jj-1]:y_dims[jj],:]

                    features.append(crop_img)
                    label.append(l)
                    immagini.append(crop_img)
                    #save image:
                    name = l + "_texture0" + str(j + 1) + "_crop_ " + str(crop_count) + ".jpg"
                    crop_count = crop_count + 1
                    name_crop = os.path.join(lichen_folder, name )
                    if(os.path.isfile(name_crop)==False):
                        is_written = cv2.imwrite(name_crop, crop_img) 

                        

                    # update df

                    df = df.append({'patch':name,'directory':name_crop,'lichen':l,'class':i}
                                   ,ignore_index = True)
        
        digit_indices[i] = immagini        
    return features,label,df,dic_label,digit_indices
                
                
                
        
        
        
        
    
    
    
    
    
    
    
    
```
